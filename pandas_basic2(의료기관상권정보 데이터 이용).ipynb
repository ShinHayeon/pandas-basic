### 무관한 여러가지 데이터를 색인 후 제거

### contains로 여러단어 검색 방법

```python
# 종합병원과 무관한 꽃배달|의료기|장례식장|상담소|어린이집 데이터 전처리 
df_seoul_hospital["상호명"].str.contains("꽃배달|의료기|장례식장|상담소|어린이집")
```

```python
df_seoul_hospital[df_seoul_hospital["상호명"].str.contains("꽃배달|의료기|장례식장|상담소|어린이집")].index
```

```python
# drop 하기 위해 list 형태로 변환
drop_row = df_seoul_hospital[df_seoul_hospital["상호명"].str.contains("꽃배달|의료기|장례식장|상담소|어린이집")].index
drop_row = drop_row.tolist()
drop_row
```

### endswith 

```python
# 의원으로 끝나는 데이터도 종합병원으로 볼 수 없기 때문에 전처리
drop_row2 = df_seoul_hospital[df_seoul_hospital["상호명"].str.endswith("의원")].index
drop_row2 = drop_row2.tolist()
drop_row2
```

```python
# 삭제할 행을 모두 합한다.
drop_row = drop_row + drop_row2
len(drop_row)
```

### 삭제 drop

```python
print(df_seoul_hospital.shape)
df_seoul_hospital = df_seoul_hospital.drop(drop_row, axis=0) # axis=0 행(인덱스번호)을 기준
print(df_seoul_hospital.shape)
```
