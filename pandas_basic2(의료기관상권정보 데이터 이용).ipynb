### 결측치 isnull

```python
df.isnull()
# true 는 null값 , false는 값이 들어있다. 
```

```python
null_count = df.isnull().sum()
#컬럼의 결측치 개수를 나타내줌 
null_count
```

### plot

```python
null_count.plot()

# 위쪽에서 아래와 같이 해주지않으면 한글이 깨져서 나온다. 
# plt.rc('font' , family= 'Malgun Gothic')
# plt.rc('axes', unicode_minus = False)
```

```python
null_count.plot.bar(rot = 60) 
```

```python
null_count.plot.barh(figsize = (5,7)) 
# figsize 로 사이즈 조정
# barh 는 horizontal 로 변경 
```

### reset_index

```python
df_null_count = null_count.reset_index()
# reset_index 로 데이터프레임을 생성 
df_null_count
```

### sort_values를 통해 결측치수를 확인

```python
df_null_count.sort_values(by = "결측치수",  ascending=False)
# 결측치로 인해 속도가 너무 오래 걸린다면 결측치 제거가 필요하다. 
```

```python
df_null_count_top = df_null_count.sort_values(by = "결측치수",  ascending=False).head(10)
# 결측치 수가 제일 많은 10개 컬럼 
```

```python
df_null_count_top["컬럼명"]
```
